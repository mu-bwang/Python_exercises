## pipe flow velocity profiles

Smooth pipe flow has the velocity distribution of:

$\frac{u}{u_*}=\frac{u_* y}{\nu}$ for $0<\frac{u_* y}{\nu}<5$

and

$\frac{u}{u_*}= \frac{1}{\kappa} \ln\left( \frac{u_*y}{\nu} \right) + 5.5$ for $20<\frac{u_* y}{\nu}<10^5$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot normalized variables
y_plus1 = np.linspace(0,10)
y_plus2 = np.linspace(10,1e5)

u_plus1 = y_plus1
kappa = 0.41
u_plus2 = 1/kappa * np.log(y_plus2) + 5.5

plt.plot(u_plus1, y_plus1)
plt.plot(u_plus2, y_plus2)
plt.grid(True)
plt.ylabel(r'$y^+$')
plt.xlabel(r'$U^+$')
plt.title('Velocity profile from the wall to center of pipe')
```

We oftern plot them in a log-linear fomrat to reveal the log-law region

```python
plt.plot(y_plus1, u_plus1)
plt.plot(y_plus2, u_plus2)
plt.grid(True)
plt.xscale('log')
plt.xlabel(r'$y^+$')
plt.ylabel(r'$U^+$')
```

## Power law approximation

Power law relation can be approximatedly used as the velocity distribution of turbulent pipe flows:

$u/u_\textrm{max} = (1-r/R)^{1/m}$, where $m$ is a function of Reynolds number

```python
r_R = np.linspace(0,1,1000)
for m in [6,8,10]:
    u_r = (1-r_R)**(1/m)
    plt.plot(u_r, r_R, label = 'm = ' + str(m))
plt.xlabel(r'$u/u_m$')
plt.ylabel(r'$r/R$')
plt.legend()
```
