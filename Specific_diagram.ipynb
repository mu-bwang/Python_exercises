EXAMPLE 5.2.2

For a rectangular channel of 20 ft width, construct a family of specific energy curves for Q = 0, 50, 100, and 300 cfs. Draw the locus of the critical depth points on these curves. For each flow rate, what is the minimum specific energy found from these curves?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# list all given parameters
Q = [0, 50, 100, 300]
W = 20
g = 32.17 # in ft/s^2

# guess a possible water depth range from 0 to 5 ft
y = np.linspace(0.1,5,100) 
```

Specific energy is defined as $E = y + \frac{V^2}{2g}$, where $V = \frac{Q}{A} = \frac{Q}{yW}$.

Critical water depth can be calculated from $y_c = \left( \frac{Q^2}{W^2g} \right)^{1/3}$.

```python
for i in range(0,len(Q)):
    E = y + (1/(2*g)) * (Q[i]/(y*W))**2
    plt.plot(E,y)
    
plt.xlabel('E (ft)')
plt.ylabel('y (ft)')
plt.axis([0,5,0,5])    
plt.legend(['Q = 0 cfs','Q = 50 cfs','Q = 100 cfs','Q = 300 cfs'])
```

```python
# calculate critical depth for Q>0
for i in range(1,len(Q)):
    E = y + (1/(2*g)) * (Q[i]/(y*W))**2
    yc = (Q[i]**2/(W**2 * g))**(1/3)
    Ec = yc + (1/(2*g)) * (Q[i]/(yc*W))**2
    plt.plot(E,y)
    plt.plot(Ec,yc,'o')
    
plt.plot([0,5],[0,5],'k--')
plt.xlabel('E (ft)')
plt.ylabel('y (ft)')
plt.axis([0,5,0,5]) 
```
